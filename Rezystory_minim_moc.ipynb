```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd
 
style = {'description_width': 'initial'}
# Slidery
Voltage1_slider = FloatSlider(min=0,max = 10, value= 6, step = 1.0, description="Napięcie1:", style= style )
Current1_slider = FloatSlider(min=0,max = 10, value= 4, step = 1.0, description="Prąd1:", style= style )
Voltage2_slider = FloatSlider(min=0,max = 10, value= 10, step = 1.0, description="Napięcie2:", style= style )
Current2_slider = FloatSlider(min=0,max = 10, value= 2, step = 1.0, description="Prąd2:", style= style )
Voltage3_slider = FloatSlider(min=0,max = 10, value= 4, step = 1.0, description="Napięcie3:", style= style )
Current3_slider = FloatSlider(min=0,max = 10, value= 2, step = 1.0, description="Prąd3:", style= style )
Voltage4_slider = FloatSlider(min=0,max = 10, value= 7, step = 1.0, description="Napięcie4:", style= style )
Current4_slider = FloatSlider(min=0,max = 10, value= 2, step = 1.0, description="Prąd4:", style= style )
Voltage5_slider = FloatSlider(min=0,max = 10, value= 3, step = 1.0, description="Napięcie5:", style= style )
Current5_slider = FloatSlider(min=0,max = 10, value= 4, step = 1.0, description="Prąd5:", style= style )     


def hurt_produkcja(NapiecieR1= 6.0,
    PradR1= 4.0,
    NapiecieR2=10,
    PradR2=2,
    NapiecieR3=4,
    PradR3=2,
    NapiecieR4=7,
    PradR4=2,
    NapiecieR5=3,
    PradR5=4,
    DI=1.0):
    
    prob = LpProblem("Problem",LpMinimize)
 
    # Zmiene określające rezystory
    x1=LpVariable("R1",0)       
    x2=LpVariable("R2",0)
    x3=LpVariable("R3",0)
    x4=LpVariable("R4",0)
    x5=LpVariable("R5",0)
    
    # Wzór na łączną moc rozproszoną na rezystorach
    prob += PradR1*PradR1*(x1) + PradR2*PradR2*(x2) + PradR3*PradR3*(x3) + PradR4*PradR4*(x4) + PradR5*PradR5*(x5), "Moc rozproszona"
   

    # Warunki
    prob += x1==NapiecieR1/PradR1, "Rezystancja1"
    prob += PradR1 >= PradR1 - DI, "Ograniczenie prądu 1"
    prob += PradR1 <= PradR1 + DI, "Ograniczenie prądu 2"
    
    prob += x2==NapiecieR2/PradR2, "Rezystancja2"
    prob += PradR2 >= PradR2 - DI, "Ograniczenie prądu 3"
    prob += PradR2 <= PradR2 + DI, "Ograniczenie prądu 4"
    
    prob += x3==NapiecieR3/PradR3, "Rezystancja3"
    prob += PradR3 >= PradR3 - DI, "Ograniczenie prądu 5"
    prob += PradR3 <= PradR3 + DI, "Ograniczenie prądu 6"
    
    prob += x4==NapiecieR4/PradR4, "Rezystancja4"
    prob += PradR4 >= PradR4 - DI, "Ograniczenie prądu 7"
    prob += PradR4 <= PradR4 + DI, "Ograniczenie prądu 8"
    
    prob += x5==NapiecieR5/PradR5, "Rezystancja5"
    prob += PradR5 >= PradR5 - DI, "Ograniczenie prądu 9"
    prob += PradR5 <= PradR5 + DI, "Ograniczenie prądu 10"
    #Koszty=4*(Ania*(a)+Stefan*(b)+Hektor*(c)+Olaf*d+Lidia*e+Piotr*f)
    prob.solve()
    # The status of the solution is printed to the screen
    print("Status:", LpStatus[prob.status])
    # Each of the variables is printed with it's resolved optimum value
    for v in prob.variables():
        print(v.name, "=", v.varValue)
    print("Minimalna moc rozproszenia {} mW  ".format( value(prob.objective)))
    
    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print("*"*10+"Shadow Prices"+"*"*10)
    print(pd.DataFrame(shadows))
    
interact(hurt_produkcja,
    NapiecieR1 = Voltage1_slider,
    PradR1 = Current1_slider,
    NapiecieR2 = Voltage2_slider,
    PradR2 = Current2_slider,
    NapiecieR3 = Voltage3_slider,
    PradR3 = Current3_slider,
    NapiecieR4 = Voltage4_slider,
    PradR4 = Current4_slider,
    NapiecieR5 = Voltage5_slider,
    PradR5 = Current5_slider)


```
