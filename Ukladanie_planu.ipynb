```python
from pulp import *
import pandas as pd
```

```python
# Układanie planu
from pulp import *

#Dane
dni_tygodnia = ['Pn','Wt','Sr','Czw','Pt']     
imiona = ["Ania",'Stefan','Hektor','Olaf','Lidia','Piotr']
ograniczenie =  {'Ania':3,
                'Stefan':3,
                'Hektor':3,
                'Olaf':3,
                'Lidia':3,
                'Piotr':3
                }     
ile_na_godzine = {'Ania':100.0,
                  'Stefan':50.0,
                  'Hektor':60.0,
                  'Olaf':40.0,
                  'Lidia':110.0,
                  'Piotr':70.0
                  }
ile_w_danym_dniu = {'Pn':2,
                    'Wt':1,
                    'Sr':1,
                    'Czw':1,
                    'Pt':3
                    } 
rob_dzien = set();
     
for pracownik in imiona:
    for dzien in dni_tygodnia:
        rob_dzien.add((pracownik,dzien))
                         
prob = LpProblem("Problem_planu", LpMinimize)        
Spr = LpVariable.dicts('',rob_dzien,cat='Binary')
# Warunki
prob += lpSum(ile_na_godzine[X[0]] * Spr[X] for X in rob_dzien)
for pracownik in imiona:
    prob += lpSum(Spr[X] for X in rob_dzien if X[0]==pracownik) <= ograniczenie[pracownik],'Warunek1%s'%pracownik
for dzien in dni_tygodnia:
    prob += lpSum(Spr[X] for X in rob_dzien if X[1]==dzien) >= ile_w_danym_dniu[dzien], 'Warunek2%s'%dzien
#prob.writeLP("Model.lp")
        
prob.solve()
print("Status:", LpStatus[prob.status])       
print('Najmniejszy koszt zatrudnienia pracowników na tydzień wynosi: '+ str(prob.objective.value()) + 'zł')      
plan = []
# Pętla której zadaniem jest ułozeniem planu  w tablicy       
for y in prob.variables():
    if y.varValue == 0:
        plan.append('Niepracuje'+ y.name )
    else:
        plan.append('Pracuje' + y.name )
print('\nUłożony plan:') 
#Pętla od wyswietlenia ułożonego planu       
for z in plan:
    print(z)
```
