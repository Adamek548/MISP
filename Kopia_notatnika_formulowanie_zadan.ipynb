W sklepie ze zdrową żywnością wytwarza się produkty dżem jabłkowy i mus jabłkowy. Wytworzenie jednostki produktu I wymaga zużycia 8 jabłek i 5 łyżek miodu, zaś jednostki musu - 2 jabłek i 5 łyżek miodu. Dostawy surowców w każdym dniu wynosza dla jabłek i miodu odpowiednio 40 i 25 jednostek. Dżem sprzedaje się po 12 zł a mus po 11 zł. Jabłko kosztuje 0.3 zł, łyżka miodu 0.2. 

Określ dzienną produkcję każdego produktu tak, aby otrzymać maksymalny zysk. Dla dwóch przypadków:

- ( HURT) wielkość produkcji można wyrazić jako nieujemne liczby rzeczywiste ( np 1.2 kg dżemu)

- (DETAL) wielkości produkcji mierzy się w jednostkach niepodzielnych ( 1 słoik dżemu). W tym przypadku do każdej wyprodukowanej jednostki trzeba doliczyć opakowanie za 0.5 zł


## Rozwiązanie

Słownie:

Zysk = Zarobek na dzemie + zarobek na musie - koszt dzemu - koszt musu - dodatkowy koszt

Zmienne decyzyjne:

Ilość ( masa lub ilosc sztuk) wyprodukowanych dzemow i musuw

Ograniczenia:

Ograniczone dostawy

Znamy ile kosztuje mus i drzem

```python
from pulp import *
import pandas as pd
```

```python
#w pulp



prob = LpProblem("Najlepsza produkcja HURT",LpMaximize)

# 2 zmienne
dzem=LpVariable("Dzem",0) # juz tutaj podajemy ograniczenie ze sa nieujemne
mus=LpVariable("Mus",0)
# pomocnicze zmienne
zuzycie_jablek =LpVariable("Jablka",0)
zuzycie_miodu =LpVariable("Miod",0)

prob += 16* dzem + 11* mus - 0.3 * zuzycie_jablek - 0.2 * zuzycie_miodu
prob += 8 * dzem+ 2 * mus == zuzycie_jablek, "Zuzycie jablek"
prob += 5 * dzem + 5 * mus == zuzycie_miodu, "Zuzycie miodu"
prob += zuzycie_jablek <= 40, "Maksimum jablek"
prob += zuzycie_miodu <= 25, "Maksimum miodu"
```

```python
prob.solve()
# The status of the solution is printed to the screen
print("Status:", LpStatus[prob.status])
# Each of the variables is printed with it's resolved optimum value
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Zysk wyniesie {} zł  ".format( value(prob.objective)))
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd
 
style = {'description_width': 'initial'}
 
 # Slidery
aluminium_slider = FloatSlider(min=0,max = 10, value= 3, step = 1.0, description="zysk_aluminium:", style= style )
czas_aluminium_F1_slider = FloatSlider(min=0,max = 10, value= 1, step = 1.0, description="czas_zrobienia_aluminium_F1:", style= style )
czas_drewno_F2_slider = FloatSlider(min=0,max = 10, value= 2, step = 1.0, description="czas_zrobienia_drewno_F2:", style= style )
czas_aluminium_F3_slider = FloatSlider(min=0,max = 10, value= 3, step = 1.0, description="czas_zrobienia_aluminium_F3:", style= style )
czas_drewno_F3_slider = FloatSlider(min=0,max = 9, value= 2, step = 1.0, description="czas_zrobienia_drewno_F3:", style= style )
max_F1_slider = FloatSlider(min=0,max = 4, value=  4, step = 1.0, description="F1_ile_produkować:", style= style )
max_F2_slider = FloatSlider(min=0,max = 12, value= 12, step = 1.0, description="F2_ile_produkować:", style= style )
max_F3_slider = FloatSlider(min=0,max = 18, value= 18, step = 1.0, description="F3_ile_produkować:", style= style )
drewno_slider = FloatSlider(min=0,max = 10, value= 5, step = 1.0, description="zysk_drewno:", style= style )
 
def hurt_produkcja(aluminium= 3.0,
    drewno= 5.0,
    czas_aluminium_F1=1,
    czas_aluminium_F3=3,
    czas_drewno_F2=2,
    czas_drewno_F3=2,
    max_F1=4,
    max_F2=12,
    max_F3=18):
    
    prob = LpProblem("Problem4",LpMaximize)
 # Zmienne
    x1=LpVariable("aluminium",0)
    x2=LpVariable("drewno",0)
 # Warunki
    prob += aluminium*(x1) + drewno*(x2), "Total zysk"
 
    prob += czas_aluminium_F1*x1 <= max_F1, "F1Requirement"
    prob += czas_drewno_F2*x2 <= max_F2, "F2Requirement"
    prob += czas_aluminium_F3*x1 + czas_drewno_F3*x2 <= max_F3, "F3Requirement"
    
    prob.solve()
    #Wyniki
    # The status of the solution is printed to the screen
    print("Status:", LpStatus[prob.status])
    # Each of the variables is printed with it's resolved optimum value
    for v in prob.variables():
        print(v.name, "=", v.varValue)
    print("Zysk będzie wynosił: {} zł  ".format( value(prob.objective)))
    
    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print("*"*10+"Shadow Prices"+"*"*10)
    print(pd.DataFrame(shadows))
    
    
interact(hurt_produkcja,
    aluminium = aluminium_slider,
    drewno = drewno_slider,
    czas_aluminium_F1 = czas_aluminium_F1_slider,
    czas_drewno_F2 = czas_drewno_F2_slider,
    czas_aluminium_F3 = czas_aluminium_F3_slider,
    czas_drewno_F3 = czas_drewno_F3_slider,
    max_F1 = max_F1_slider,
    max_F2 = max_F2_slider,
    max_F3 = max_F3_slider)
```
